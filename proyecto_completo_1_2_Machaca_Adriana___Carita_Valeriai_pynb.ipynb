## **4. Marco Conceptual**

- **Media:**  
Es el promedio aritmético de un conjunto de valores numéricos, calculado sumando todos los valores y dividiendo entre el número total de elementos. La media es una medida de tendencia central que permite identificar un valor representativo dentro de un conjunto de datos, útil para analizar precios promedio, cantidad vendida o ticket promedio en un eCommerce. Permite detectar desviaciones respecto al comportamiento típico de las ventas y comparar rendimientos entre categorías o países. (Montgomery, 2017)

- **Varianza:**  
La varianza mide la dispersión de los datos alrededor de la media, indicando cuánto se alejan los valores individuales del promedio. En un contexto de eCommerce, la varianza de precios o cantidades vendidas ayuda a identificar inconsistencias o productos que tienen ventas muy fluctuantes. Es fundamental para calcular otras métricas estadísticas como la desviación estándar y para detectar patrones inusuales en los datos. (Keller, 2018)

- **Outlier (Valor atípico):**  
Un outlier es un valor que difiere significativamente del resto del conjunto de datos. En análisis de ventas, un outlier puede ser una transacción con un precio extremadamente alto o bajo, o una cantidad vendida inusual. Detectar outliers es importante porque estos valores pueden distorsionar promedios, varianzas y otras métricas, y en ocasiones representan errores de registro o situaciones especiales que requieren revisión. (Field, 2013)

- **Segmentación:**  
La segmentación consiste en dividir un conjunto de datos en grupos homogéneos según criterios específicos, como país, categoría de producto, rango de precios o comportamiento de compra. Permite un análisis más preciso del comportamiento de clientes y ventas, identificar nichos de mercado, ajustar campañas de marketing y optimizar la gestión de inventario según patrones específicos detectados en cada segmento. (Hair et al., 2014)

- **Normalización de texto:**  
Es el proceso de ajustar cadenas de texto a un formato uniforme, generalmente convirtiendo todas las letras a minúsculas y eliminando espacios innecesarios al inicio o al final. En análisis de eCommerce, esto garantiza que las categorías de productos, nombres de productos y países se agrupen correctamente, evitando duplicados que solo difieren por mayúsculas o espacios. Esto es clave para contar correctamente transacciones, generar reportes precisos y realizar análisis categóricos confiables. (Manning et al., 2008)

## **10. Limpieza del Dataset**


```python
# =========================
# 1. Valores nulos
# =========================
ventas_df['fecha'] = pd.to_datetime(ventas_df['fecha'], errors='coerce')
ventas_df['precio'] = ventas_df['precio'].fillna(ventas_df['precio'].mean())
ventas_df['cantidad'] = ventas_df['cantidad'].fillna(1)
ventas_df['producto'] = ventas_df['producto'].fillna('desconocido')
ventas_df['categoria'] = ventas_df['categoria'].fillna('desconocido')
ventas_df['pais'] = ventas_df['pais'].fillna('desconocido')
ventas_df['reembolsos'] = ventas_df['reembolsos'].fillna('no')

# =========================
# 2. Duplicados
# =========================
ventas_df = ventas_df.drop_duplicates()

# =========================
# 3. Conversión de tipos
# =========================
ventas_df['precio'] = ventas_df['precio'].astype(float)
ventas_df['cantidad'] = ventas_df['cantidad'].astype(int)
ventas_df['producto'] = ventas_df['producto'].astype(str)
ventas_df['categoria'] = ventas_df['categoria'].astype(str)
ventas_df['pais'] = ventas_df['pais'].astype(str)
ventas_df['reembolsos'] = ventas_df['reembolsos'].map({'no':0,'si':1})

# =========================
# 4. Rangos válidos
# =========================
ventas_df = ventas_df[(ventas_df['precio'] > 0) & (ventas_df['cantidad'] > 0)]
ventas_df = ventas_df[ventas_df['reembolsos'].isin([0,1])]

# =========================
# 5. Normalización de texto
# =========================
ventas_df['producto'] = ventas_df['producto'].str.lower().str.strip()
ventas_df['categoria'] = ventas_df['categoria'].str.lower().str.strip()
ventas_df['pais'] = ventas_df['pais'].str.lower().str.strip()

# =========================
# Verificar limpieza final
# =========================
print("\nPrimeras filas después de limpieza:")
display(ventas_df.head(10))

print("\nInformación del dataset después de limpieza:")
ventas_df.info()
```

## **11. Transformaciones y limpieza aplicada — decisiones ejecutadas**

**Conversión de tipos:**
La columna fecha se convirtió a datetime (errores → NaT) para permitir análisis temporal; precio a float; cantidad y reembolsos a int/binario; texto a string.

**Valores nulos:**

**fecha:** se eliminaron filas sin fecha (poca utilidad sin fecha).

**precio:** imputado con la media de la columna cuando el % de nulos es bajo.

**cantidad:** imputado con 1 cuando faltaba (asumimos que faltó registro de cantidad).

producto, categoria, pais: rellenados con 'desconocido'.

**reembolsos:** normalizado a 0/1 (map no→0, si→1).

**Duplicados:** eliminación de filas duplicadas exactas con drop_duplicates().

**Rangos válidos:** se filtraron filas con precio <= 0 o cantidad <= 0 por ser registros no válidos para métricas de venta.

**Normalización de texto:** lower().strip() en columnas categóricas para evitar duplicados por mayúsculas/espacios.

**Justificación corta:** Estas decisiones buscan asegurar consistencia en tipos para operaciones numéricas/temporales, evitar sesgos por nulos extremos y eliminar registros inválidos que distorsionen métricas (ingresos, tickets, promedios).

```python
# Celda: Limpieza reproducible basada en las decisiones anteriores
# (ejecuta esto después de cargar si no lo hiciste)
df = ventas_df.copy()

# Conversión de tipos y limpieza inicial
df['fecha'] = pd.to_datetime(df['fecha'], errors='coerce')
df['precio'] = pd.to_numeric(df['precio'], errors='coerce')
df['cantidad'] = pd.to_numeric(df['cantidad'], errors='coerce')
# Normalizar texto
for c in ['producto','categoria','pais']:
    if c in df.columns:
        df[c] = df[c].astype(str).str.lower().str.strip().replace({'nan':'desconocido'})

# Imputaciones simples
if 'precio' in df.columns:
    df['precio'] = df['precio'].fillna(df['precio'].mean())
if 'cantidad' in df.columns:
    df['cantidad'] = df['cantidad'].fillna(1).astype(int)
for c in ['producto','categoria','pais']:
    df[c] = df[c].fillna('desconocido')

# Reembolsos -> mapear si/ no a 1/0, y forzar int
if 'reembolsos' in df.columns:
    df['reembolsos'] = df['reembolsos'].map({'si':1,'sí':1,'yes':1,'no':0, '0':0, '1':1}).fillna(0).astype(int)

# Eliminar filas sin fecha (si existen)
df = df[~df['fecha'].isna()]

# Rangos válidos
df = df[(df['precio'] > 0) & (df['cantidad'] > 0)]

# Eliminar duplicados exactos
df = df.drop_duplicates().reset_index(drop=True)

# Columna ingreso
df['ingreso'] = df['precio'] * df['cantidad']

# Verificación rápida
print("Dimensiones tras limpieza:", df.shape)
display(df.head())
```

##**¿Qué categorías de productos generan mayores ingresos?**

**Hipótesis:** Productos tecnológicos tienen mayores ventas promedio.



```python
import matplotlib.pyplot as plt

ventas_df['ingreso'] = ventas_df['precio'] * ventas_df['cantidad']

ingresos_categoria = ventas_df.groupby('categoria')['ingreso'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
ingresos_categoria.plot(kind='bar')
plt.title("Ingresos totales por categoría")
plt.xlabel("Categoría")
plt.ylabel("Ingresos")
plt.tight_layout()
plt.show()
```

##**¿Qué países presentan más devoluciones?**

**Hipótesis:** Países con más pedidos registran más reembolsos.



```python
import matplotlib.pyplot as plt

# Total de devoluciones por país
devoluciones_pais = ventas_df.groupby('pais')['reembolsos'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
plt.barh(devoluciones_pais.index, devoluciones_pais.values, color='salmon')
plt.title('Total de devoluciones por país')
plt.xlabel('Número de devoluciones')
plt.ylabel('País')
plt.gca().invert_yaxis()  # Para mostrar el más alto arriba
plt.show()
```

**Análisis**

*Los países con más pedidos tienen más devoluciones totales, pero algunos países pequeños muestran alta proporción de devoluciones por pedido, indicando diferencias en satisfacción o logística.*


##**¿En qué meses o fechas se concentran más ventas?**
**Hipótesis:** Ventas aumentan en meses festivos .


```python
# Crear columna de mes
ventas_df['mes'] = ventas_df['fecha'].dt.month

# Ingreso total por mes
ingreso_mes = ventas_df.groupby('mes')['precio'].sum()

plt.figure(figsize=(10,6))
plt.plot(ingreso_mes.index, ingreso_mes.values, marker='o')
plt.title('Ingreso total por mes')
plt.xlabel('Mes')
plt.ylabel('Ingreso total')
plt.xticks(range(1,13))
plt.grid(True)
plt.show()
```

**Análisis:**

##**14. Feature engineering básico**

ingreso = precio * cantidad.

Justificación: métrica primaria de negocio para medir facturación por transacción.

year, month, month_name y dayofweek extraídos de fecha.

Justificación: analizar estacionalidad y patrones temporales.

rango_precio (bins por cuantiles: bajo/medio/alto).

Justificación: agrupar productos por tramo de precio para comparar comportamiento de volumen y reembolsos entre rangos.

cantidad_segment (bajo/medio/alto) usando cuantiles.

Justificación: identificar productos de volumen alto/bajo y su impacto en ingresos y devoluciones.

```python
# Celda: Feature engineering
df['year'] = df['fecha'].dt.year
df['month'] = df['fecha'].dt.month
df['month_name'] = df['fecha'].dt.strftime('%Y-%m')
df['dayofweek'] = df['fecha'].dt.day_name()

# Rangos de precio por cuantiles (bajo/medio/alto)
q1 = df['precio'].quantile(0.33)
q2 = df['precio'].quantile(0.66)
def precio_tramo(x):
    if x <= q1:
        return 'bajo'
    elif x <= q2:
        return 'medio'
    else:
        return 'alto'
df['rango_precio'] = df['precio'].apply(precio_tramo)

# Segmento de cantidad por cuantiles
c1 = df['cantidad'].quantile(0.33)
c2 = df['cantidad'].quantile(0.66)
def cantidad_tramo(x):
    if x <= c1:
        return 'baja'
    elif x <= c2:
        return 'media'
    else:
        return 'alta'
df['cantidad_segment'] = df['cantidad'].apply(cantidad_tramo)

# Mostrar ejemplos
display(df[['precio','rango_precio','cantidad','cantidad_segment']].head(10))
```

# **18. Conclusiones y recomendaciones**

### **Conclusiones**
1. Las categorías con mayor ingreso son estratégicas y deben priorizarse en inventario y campañas.  
2. Los países presentan diferencias: algunos generan alto ticket promedio y otros alto volumen de ventas.  
3. Se identifican picos estacionales en ciertos meses, útiles para planificación de promociones y stock.  
4. Productos de precio bajo se venden más en volumen; los de precio alto generan más ingreso por unidad.  
5. Algunas categorías muestran alta tasa de reembolsos, indicando posibles problemas de calidad o descripción.

### **Recomendaciones accionables**
- Priorizar el Top 10 de categorías para campañas y control de inventario.  
- Segmentar estrategias según mercado: volumen vs ticket promedio.  
- Preparar inventario y promociones anticipadas en meses de alta demanda.  
- Revisar categorías con alta tasa de reembolsos para mejorar satisfacción del cliente.  
- Usar tramos de precio y cantidad para diseñar estrategias de descuentos y marketing.
